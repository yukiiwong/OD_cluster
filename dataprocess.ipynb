# 数据预处理阶段
## 1.导入数据
导入手机信令数据

```python
import pandas as pd
import numpy as np
# 数据处理
#读取数据
#数据精简 随机不重复抽取10000行数据样本
data_ori = pd.read_excel('16_09_02-09_07.xls').sample(10000)
```

```python
data_ori.head(5)
```

## 2.数据筛选
将限定时间范围内的数据筛选


```python
#筛选出ACCESS_TIME中包含'06-SEP-16'的数据
#将数据根据每个人ID排序
data = data_ori[data_ori['ACCESS_TIME'].str.contains('10-SEP-16')].sort_values(by='IMSI')
```

```python
data.head(15)
```

## 3.数据清洗

```python
#计算将仅出现一次的人剔除
#将重复值的INDEX提取出来
data_bool = data.duplicated(subset=['IMSI'])
rep = []
for i in range(len(data_bool)):
    if data_bool.iloc[i] == False:
        rep.append(i)
#记录OD
rep.append(3.1415926)
origin = []
dest = []
for i in range(len(rep)):
    #判断该用户是否存在多次信息收集，是则收集OD
    if rep[i] != 3.1415926:
        if rep[i]+1 < rep[i+1]:
            d = rep[i]
            #收集该用户的多段OD
            while (d < rep[i+1]):
                origin.append([data.iloc[d,0],data.iloc[d,5],data.iloc[d,6]])
                dest.append([data.iloc[d+1,5],data.iloc[d+1,6]])
                d += 1
    else:
        if rep[i-1] + 1 < len(data):
            d = rep[i-1]
            while (d < len(data)):
                origin.append([data.iloc[d,0],data.iloc[d,5],data.iloc[d,6]])
                dest.append([data.iloc[d+1,5],data.iloc[d+1,6]])
                d += 1
```

## 4.储存数据

```python
start_frame = pd.DataFrame(origin,columns=['IMSI','o_longitude','o_latitude'])
end_frame = pd.DataFrame(dest,columns=['d_longitude','d_latitude'])
GPS_frame = pd.concat([start_frame,end_frame],axis=1)
GPS_frame.to_csv('trajectory_res.csv',index=False)
```
